```python
from scipy.stats import pearsonr
def calculate_pvalues(df):
    df = df.select_dtypes(include=['number'])
    pairs = pd.MultiIndex.from_product([df.columns, df.columns])
    pvalues = [pearsonr(df[a], df[b])[1] for a, b in pairs]
    pvalues = pd.Series(pvalues, index=pairs).unstack().round(4)
    return pvalues
```

Doing the correlation matrix for all people:

```python
all_data = pd.read_excel("C:/Users/sarah/Desktop/W2023 MoCap Data/all_data.xlsx")
```

```python
all_data = all_data.drop(columns=['Observations', 'handedness', 'max_hand_used'])
```

```python
from sklearn.preprocessing import StandardScaler
z_scaler = StandardScaler()
z_scaler = z_scaler.set_output(transform="pandas")
#z = z_scaler.fit(P1_data)
all_data_norm = z_scaler.fit_transform(all_data)
```

```python
corr_matrix_pearson = all_data_norm.corr(method="pearson")
```

```python
corr_matrix_pearson
```

```python
from scipy.stats import pearsonr
def calculate_pvalues(df):
    df = df.select_dtypes(include=['number'])
    pairs = pd.MultiIndex.from_product([df.columns, df.columns])
    pvalues = [pearsonr(df[a], df[b])[1] for a, b in pairs]
    pvalues = pd.Series(pvalues, index=pairs).unstack().round(4)
    return pvalues
```

```python
pvalues_pearson = calculate_pvalues(all_data_norm)
```

```python
pvalues_pearson
```

```python
corr_matrix_spearman = all_data_norm.corr(method="spearman")
```

```python
corr_matrix_spearman
```

```python
fig, ax = plt.subplots(figsize=(13,13))
sb.heatmap(corr_matrix_spearman, annot=True)
plt.show()
plt.savefig('corr_matrix_spearman')
```

```python
corr_matrix_spearman.to_excel('corr_matrix_spearman.xlsx')
```

```python
from scipy.stats import spearmanr
def calculate_pvalues_spear(df):
    df = df.select_dtypes(include=['number'])
    pairs = pd.MultiIndex.from_product([df.columns, df.columns])
    pvalues = [spearmanr(df[a], df[b])[1] for a, b in pairs]
    pvalues = pd.Series(pvalues, index=pairs).unstack().round(4)
    return pvalues
```

```python
pvalues_spearman = calculate_pvalues_spear(all_data_norm)
```

```python
pvalues_spearman
```

```python
pvalues_spearman.to_excel('SpearmanRho_pvalues.xlsx')
```
